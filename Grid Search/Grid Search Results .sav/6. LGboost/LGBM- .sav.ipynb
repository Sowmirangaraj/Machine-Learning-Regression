```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import lightgbm as lgb
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset.columns
```

```python
indep=dataset[['age','bmi','children','sex_male', 'smoker_yes']]
```

```python
dep=dataset[['charges']]
```

```python
from lightgbm import LGBMRegressor
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'num_leaves' : [60],
               'learning_rate' : [0.1,0.3],
                'num_iterations': [50,100],
                'max_depth': [5],
                'min_data_in_leaf': [75],
                'bagging_fraction': [0.9],
                'feature_fraction': [0.8],
                'n_estimators':[10,100]}
grid = GridSearchCV(LGBMRegressor(), param_grid, refit = True, verbose = 3,n_jobs=-1)
grid.fit(indep,dep)
```

```python
import pickle
filename="finalized_model_LGBM.sav"
pickle.dump(grid,open(filename,'wb'))
```

```python
loaded_model=pickle.load(open("finalized_model_LGBM.sav",'rb'))
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_Prediction={}".format(Future_Prediction))
 
```
